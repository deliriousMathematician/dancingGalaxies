```python
import pynbody as pyn
import matplotlib.pyplot as plt
import matplotlib.animation as ani
import numpy as np
```

```python
sim = pyn.load('simFiles/run708main.01000')
```

```python
sim.physical_units()
sim.s['age'].convert_units('Gyr')
```

```python
pyn.analysis.angmom.faceon(sim)
```

```python
pyn.plot.sph.image(sim.g, width=16, qty='rho', cmap=plt.cm.turbo, av_z=False)
plt.clim(10**5, 10**10)
```

```python
# Accessing the z-component of the positions of every gas particle
sim.g['pos'][:, 2]
```

```python
# Updating the z-components
sim.g['pos'][:, 2] = sim.g['pos'][:, 2] - 0.1 
```

```python
sim.g['pos'][:, 2]
```

```python
pyn.plot.sph.image(sim.g, width=16, qty='rho', cmap=plt.cm.turbo, av_z=False)
plt.clim(10**5, 10**10)
```

```python
sim2 = pyn.load('simFiles/run708main.01000')
```

```python
sim2.physical_units()
sim2.s['age'].convert_units('Gyr')
```

```python
pyn.analysis.angmom.faceon(sim2)
```

```python
plot = pyn.plot.sph.image(sim2.g, width=16, qty='rho', cmap=plt.cm.turbo, av_z=False)
plt.clim(10**5, 10**10)
plt.close()
```

```python
diff = 0.01 # distance z is shifted per frame
def update(frame):
    sim2.g['pos'][:, 2] = sim2.g['pos'][:, 2] - diff
    return plot
```

```python
ani = ani.FuncAnimation(func=update, frames=100, interval=200)
```
